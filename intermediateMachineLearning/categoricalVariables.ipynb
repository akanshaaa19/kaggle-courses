## Categorical Variables

* A lot of time we have data which only has a few options, and others have to select from that options only.
* This will give an error we we feed the data without processing
* There are threee approaches to process categorical data and make them ready for models 
#### 1) Drop Categorical column 
* This will only work if the column doesnt contain useful information
#### 2) Ordinal Encoding
* This assigns a special value to each option
* Example: Yes -> 0    No -> 1    Maybe -> 3
####  3) One-Hot Encoding
* One-hot encoding creates new columns indicating the presence (or absence) of each possible value in the original data. To understand this, we'll work through an example.
![hot encoding](assets\TW5m0aJ.png")


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
data=pd.read_csv(r"C:\Users\DELL\Desktop\Repo\Machine Learning\kaggle\datasets\melb_data.csv")
```

```python
data.head()
```

```python
y=data['Price']
X=data.drop(['Price'], axis=1)
```

```python
X.head()
```

```python
X.info()
```

```python
X_train_full, X_val_full, y_train, y_val=train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
cols_missing=[col for col in X_train_full.columns if X_train_full[col].isnull().any()]
print(cols_missing)
```

```python
X_train_full.drop(cols_missing, axis=1, inplace=True)
```

```python
X_val_full.drop(cols_missing, axis=1, inplace=True)
```

```python
X_train_full.columns
```

```python
print(X_train_full[['Type', 'Method', 'Regionname']].head(10)) #the columns which have categotical data
```

```python
X_train_full['Regionname'].nunique()
```

```python
# "Cardinality" means the number of unique values in a column
# Select categorical columns with relatively low cardinality (convenient but arbitrary)
low_cardinality_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and X_train_full[cname].dtype == "object"]
low_cardinality_cols
#nunique() r-> returns number of unique elements
```

```python
numerical_colums=[col for col in X_train_full.columns if X_train_full[col].dtype in ['float64', 'int64']]
numerical_colums
```

```python
feature_columns=numerical_colums+low_cardinality_cols
feature_columns
```

```python
X_train = X_train_full[feature_columns].copy()
X_valid = X_val_full[feature_columns].copy()
X_train.head()
```

```python
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

##### Approach one: Dropping columns

```python
X_train_dropped=X_train.select_dtypes(exclude=['object'])
X_val_dropped=X_valid.select_dtypes(exclude=['object'])
X_train_dropped.head()
```

```python
from sklearn.metrics import mean_absolute_error
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
print("MAE from Approach 1 (Dropping columns):")
score_dataset(X_train_dropped, X_val_dropped, y_train, y_val)
```

##### Approach two: Original Encoding

```python
X_train_enc=X_train.copy()
X_valid_enc=X_valid.copy()
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

```python
X_train_enc[object_cols]=enc.fit_transform(X_train[object_cols])
X_valid_enc[object_cols] = enc.transform(X_valid[object_cols])
```

```python
print("MAE from Approach 2 (Ordinal Encoding):")
print(score_dataset(X_train_enc, X_valid_enc, y_train, y_val))
```

In the code cell above, for each column, we randomly assign each unique value to a different integer. This is a common approach that is simpler than providing custom labels; however, we can expect an additional boost in performance if we provide better-informed labels for all ordinal variables.

##### Approach three: One-Hot encoding

* We set handle_unknown='ignore' to avoid errors when the validation data contains classes that aren't represented in the training data, and
* setting sparse=False ensures that the encoded columns are returned as a numpy array (instead of a sparse matrix).

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc=OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
train_oh_cols=pd.DataFrame(oh_enc.fit_transform(X_train[object_cols]))
valid_oh_cols=pd.DataFrame(oh_enc.transform(X_valid[object_cols]))
# Apply one-hot encoder to each column with categorical data
```

```python
train_oh_cols
```

```python
train_oh_cols.index=X_train.index
valid_oh_cols.index = X_valid.index
```

```python
#Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(object_cols, axis=1)
num_X_valid = X_valid.drop(object_cols, axis=1)
```

```python
X_train_oh = pd.concat([num_X_train, train_oh_cols], axis=1)
X_valid_oh = pd.concat([num_X_valid, valid_oh_cols], axis=1)
```

```python
X_train_oh
```

```python
X_train_oh.columns=X_train_oh.columns.astype(str)
X_valid_oh.columns=X_valid_oh.columns.astype(str)
#the column names are in string so changed that
```

```python
print("MAE from Approach 3 (One-hot Encoding):")
print(score_dataset(X_train_oh, X_valid_oh, y_train, y_val))
```
